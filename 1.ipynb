## Problem 1: Smart Calculator with History
**Goal**: Create a calculator that remembers the last 5 calculations.
**Key Concepts**:
- `eval()`: Evaluates mathematical strings.
- `unshift()`: Adds to the beginning of an array.
- `pop()`: Removes from the end of an array.

```python
let history = []; 
function calculateResult() {
    const display = document.getElementById('display');
    const result = eval(display.value);
    display.value = result;
    addToHistory(display.value + " = " + result);
}
function addToHistory(entry) {
    history.unshift(entry);
    if (history.length > 5) history.pop();
}
```

## Problem 3: Dynamic To-Do List
**Goal**: Add, remove, and filter tasks.
**Key Concepts**:
- `Date.now()`: Generates a unique ID.
- `filter()`: Creates a new array with only the items we want (used for deleting).
- `!task.completed`: Toggles boolean values.

```python
let tasks = [];
function addTask() {
    tasks.push({ id: Date.now(), text: "New Task", completed: false });
}
function deleteTask(id) {
    tasks = tasks.filter(task => task.id !== id);
}
```
